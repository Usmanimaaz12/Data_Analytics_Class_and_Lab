```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
path="/content/AirPassengers.csv"
df=pd.read_csv(path,parse_dates=["Date"], index_col="Date")
```

```python
plt.figure()
plt.plot(df['Passengers'], color='tab:red')
plt.show()
```
